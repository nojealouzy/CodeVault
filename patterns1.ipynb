Square of side 'N'
Problem Description: You are given an integer n. Your task is to return a square pattern of size n x n made up of the character '*', represented as a list of strings.



Input Parameters:

n (int): The size of the square (number of rows and columns).



Output:

A list of strings where each string is a row of n characters.



Example:

Input: 3
Output: ['***', '***', '***']
 
Input: 5
Output: ['*****', '*****', '*****', '*****', '*****']

```python
def generate_square(n):
    """
    Function to return a square pattern of '*' of side n as a list of strings.
    
    Parameters:
    n (int): The size of the square.
    
    
    Returns:
    list: A list of strings where each string represents a row of the square.
    """
    # Your code here
    n = n
    x = []
    mult = n*'*'
    for i in range(0,n):
        x.append(mult)
    return x
```

```python
def generate_hollow_square(n):
    """
    Function to return a hollow square pattern of '*' of side n as a list of strings.
    
    Parameters:
    n (int): The size of the square.
    
    Returns:
    list: A list of strings where each string represents a row of the hollow square.
    """
    # Your code here
    n = n
    l = []
    x = n - 2
    a = n*'*'
    if n==1:
        l.append(n*'*')

    elif n>1:
        if x>1:
            b = (int(x/x)*'*')+x*' '+(int(x/x)*'*')
            l.append(a)
            for i in range(0,x):
                l.append(b)
            l.append(a)

        else:
            b = (x*'*'+x*' '+x*'*')
            l.append(a)
            for i in range(0,x):
                l.append(b)
            l.append(a)

    return l
```

```python
def generate_triangle(n):
    """
    Function to return a right-angled triangle of '*' of side n as a list of strings.
    
    Parameters:
    n (int): The height and base of the triangle.
    
    Returns:
    list: A list of strings where each string represents a row of the triangle.
    """
    # Your code here
    n = n
    l = []
    
    no = 1
    for i in range(0,n):
        star = '*'*no
        l.append(star)
        no += 1
    return l
```

```python
def generate_inverted_triangle(n):
    """
    Function to return an inverted right-angled triangle of '*' of side n as a list of strings.
    
    Parameters:
    n (int): The height and base of the triangle.
    
    Returns:
    list: A list of strings where each string represents a row of the triangle.
    """
    # Your code here
    n = n

    l = []
    no = n
    
    if n==1:
        l.append('*'*n)
    if n>1:
        for i in range(0,n):

            star = '*'*no
            l.append(star)
            no -= 1

    return l
```

```python
def generate_pyramid(n):
    n = n
    l = []
    no = 1
    space = (n-1) #5-1 = 4
    for i in range(0,n):
        each_side = int(space)*' ' #4*
        star = no*'*'
        res = each_side+star+each_side #    *    
        l.append(res)
        space -=1
        no += 2
    return l
```

```python
x = generate_pyramid(1)
print(x)
```

```python
def generate_inverted_pyramid(n):
    n = n
    l = []
    
    no = n+(n-1)
    space = n-n
    for i in range(0,n):
        star = no*'*'
        each_space = int(space/2)*' '
        res = each_space+star+each_space
        no -= 2
        space += 2
        l.append(res)
    return l
   
```

```python
x = generate_inverted_pyramid(3)
x
```

```python
def generate_number_triangle(n):
    n = n
    l = []
    counter = 1
    letter = 1
   
  
    for i in range(n):
        res = counter * str(letter)
        l.append(" ".join(res))
        counter+=1
        letter+=1
    return l
```

```python
def generate_floyds_triangle(n):
    n=n
    l = []
    num = 1

    for i in range(1,n+1):
        row = []
        for j in range(i):
            row.append(str(num))
            num += 1
        l.append(" ".join(row))
    return l
```

```python
def generate_diamond_pattern(n):
    n = n
    l = []
    num = 1

    for i in range(1,n+1):
        row = []
        for j in range(i):
            row.append(num*'*')
            num += 2
        l.append(" ".join(row))
    return l
```

```python
def prime_checker(n):
    n=n
    is_prime = True

    if n<2:
        print(f"{n} is not a prime number")
    else:

        for i in range(2,n):

            if n%i == 0:
                is_prime = False
                break
        
        if is_prime:
            print(f"{n} is a prime number.")
        else:
            print(f"{n} is not a prime number.")
```

```python
def fibonnaci_number(n):
    """This code only prints the nth fibonacci number"""
    n = n #10
    first_no = 0
    second_no = 1

    if n==0:
        return 0
    elif n==1:
        return 1
    for i in range(0,n):
        current_no = first_no + second_no
        first_no = second_no
        second_no = current_no
    return current_no
```

```python
x = fibonnaci_number(5)
x
```
